##Com o dataframe pronto, agora iremos criar as funções de plotagem dos gráficos:

```python
def plota_histograma():
  '''função para representarmos a distribuição dos valores de uma coluna'''
  fig, ax = plt.subplots()
  ax.hist(cinco_criptos['Open'], bins=20)
  plt.xlabel('Open')

  # Incorpora a figura em um widget Tkinter
  canvas = FigureCanvasTkAgg(fig, master=janela)
  canvas_widget = canvas.get_tk_widget()
    
  # Coloca o widget na janela
  canvas_widget.grid(column=0, row=2, padx=10, pady=5)
  return
```

```python
plota_histograma()
```

```python
#para a correlação na forma de um gráfico de calor temos:
def correlacao():
    x = cinco_criptos.copy()
    x = x.drop('Name', axis=1)
    x = x.drop('Date', axis=1)
    correlacao = x.corr()

    # Criar uma nova figura e um subplot para o heatmap
    fig = plt.figure(figsize=(5, 4))
    ax = fig.add_subplot(111)

    # Plotar o heatmap no subplot
    sns.heatmap(correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1, ax=ax)

    # Configurar os limites do gráfico
    plt.ylim(len(correlacao), 0)
    plt.xlim(0, len(correlacao))

    # Incorporar a figura em um widget Tkinter
    canvas = FigureCanvasTkAgg(fig, master=janela)
    canvas_widget = canvas.get_tk_widget()

    # Colocar o widget na janela
    canvas_widget.grid(column=0, row=2, padx=10, pady=5)
```

```python
correlacao()  #os valores apresentam uma alta correlação linear.
```

```python
def fechamento_temporal():
  '''Gráfico temporal do fechamento por moeda'''
  intervalo = cinco_criptos['Date']
  fig, ax = plt.subplots()
  for moeda, grupo in cinco_criptos.groupby('Name'):
    plt.plot(grupo['Date'], grupo['Close'], label=moeda)
  plt.xlabel('Data')
  plt.ylabel('Valor de Fechamento')
  plt.gca().xaxis.set_major_locator(plt.MaxNLocator(6)) #limita o n. de rotulos no eixo X à apenas 6
  ax.legend()
  plt.title('Fechamento X Data')
  plt.show()
  
  # Incorpora a figura em um widget Tkinter
  canvas = FigureCanvasTkAgg(fig, master=janela)
  canvas_widget = canvas.get_tk_widget()
    
  # Coloca o widget na janela
  canvas_widget.grid(column=0, row=2, padx=10, pady=5)
  return
```

####Podemos perceber que as moedas Bitcoin, Ethereum e Litecoin são as mais valorizadas nesse período. <br>
Nenhuma das outras tem papel relevante no gráfico de fechamento.

```python
fechamento_temporal()
```

```python
def volume_temporal():
  '''Gráfico temporal do volume movimentado por moeda'''
  intervalo = cinco_criptos['Date']
  fig, ax = plt.subplots()
  for moeda, grupo in cinco_criptos.groupby('Name'):
    plt.plot(grupo['Date'], grupo['Volume'], label=moeda)
  plt.xlabel('Data')
  plt.ylabel('Volume Movimentado')
  plt.gca().xaxis.set_major_locator(plt.MaxNLocator(6)) #limita o n. de rotulos no eixo X à apenas 6
  ax.legend()
  plt.title('Volume X Data')
  plt.show()
  
  # Incorpora a figura em um widget Tkinter
  canvas = FigureCanvasTkAgg(fig, master=janela)
  canvas_widget = canvas.get_tk_widget()
    
  # Coloca o widget na janela
  canvas_widget.grid(column=0, row=2, padx=10, pady=5)
  return
```

###O gráfico de 'Volume Movimentado X Data' é mais volátil que o preço das criptos, porém apresenta um comportamento semelhante do valor de fechamento.  <br>
Com Bitcoin tendo o maior volume, depois Ethereum e Litecoin, e as outras criptos com pouca representatividade. <br>
(Por isso é importante observarmos a correlação).

```python
volume_temporal()
```

```python
#Gráfico de dispersão:
def dispersao():
  '''Plotar Gráfico de dispersão do volume de movimentação pelo valor de fechamento, de cada cripto.'''
  #plotando o gráfico:
  grupos_moedas = cinco_criptos.groupby('Name')

  fig, ax = plt.subplots()

  tamanho = cinco_criptos['Close'] / cinco_criptos['Close'].max()
  for moeda, grupo in grupos_moedas:
      ax.scatter(grupo['Volume'], grupo['Close'], s=(grupo['Close']/100), alpha=0.5, label=moeda)

  ax.set_xlabel('Volume')
  ax.set_ylabel('Fechamento')
  ax.set_title('Gráfico de Dispersão de Volume X Fechamento')

  ax.legend()
  plt.show()
  
  # Incorpora a figura em um widget Tkinter
  canvas = FigureCanvasTkAgg(fig, master=janela)
  canvas_widget = canvas.get_tk_widget()
    
  # Coloca o widget na janela
  canvas_widget.grid(column=0, row=2, padx=10, pady=5)
  return
```

###Observamos que a presença do Bitcoin é monstruosa se comparada com as outras criptos, e por isso assumimos que é uma cripto altamente valorizada.
Novamente, temos Ethereum em segundo e e Litecoin em terceiro. <br>
As outras criptos não apresentam volume e valor de fechamento relevantes.

```python
dispersao()
```

#Criando um ambiente interativo para a observação dos gráficos:

```python
#criando a janela principal

janela.title('Análise Anual - 5 Criptomoedas')
janela.geometry('800x800')

texto_orientacao = Label(janela, text='Esolha uma das análises') #aparecera sobre a janela
texto_orientacao.grid(column=0, row=0, padx=10, pady=5)


botao1 = Button(janela, text='Distribuição dos valores (para cada coluna)', command=plota_histograma)
botao1.grid(column=0, row=3, padx=10, pady=5)


botao2 = Button(janela, text='Correlação das Colunas', command=correlacao)
botao2.grid(column=0, row=5, padx=10, pady=5)



botao3 = Button(janela, text='Valor de Fechamento X Data', command=fechamento_temporal)
botao3.grid(column=0, row=7, padx=10, pady=5)



botao4 = Button(janela, text='Volume de Movimentação X Data', command=volume_temporal)
botao4.grid(column=0, row=9, padx=10, pady=5)


botao5 = Button(janela, text='Volume de Movimentação X Valor de Fechamento', command=dispersao)
botao5.grid(column=0, row=9, padx=10, pady=5)


janela.mainloop() #para a janela permanecer aparecendo
```
